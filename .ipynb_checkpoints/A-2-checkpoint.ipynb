```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop(df[df['math score']=='?'].index)
df = df.drop(df[df['math score']<'0'].index)
```

```python
df['math score'] = df['math score'].fillna(method = 'ffill')
df['reading score'] = df['reading score'].fillna(method = 'ffill')
df['writing score'] = df['writing score'].fillna(method = 'ffill')
```

```python
df['math score'] = df['math score'].astype('int64')
df['reading score'] = df['reading score'].astype('int64')
df['writing score'] = df['writing score'].astype('int64')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
fig , axs = plt.subplots(1,3,figsize = (15,12))

sns.boxplot(data = df['math score'], ax = axs[0], color = 'r')
sns.boxplot(data = df['reading score'], ax = axs[1], color = 'g')
sns.boxplot(data = df['writing score'], ax = axs[2], color = 'b')

plt.tight_layout()
plt.show()
```

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3-Q1
lower = Q1- 1.5*IQR
upper = Q3 + 1.5*IQR
df = df[(df['math score']> lower) & (df['math score'] < upper)]
```

```python
Q1 = df['reading score'].quantile(0.25)
Q3 = df['reading score'].quantile(0.75)
IQR = Q3-Q1
lower = Q1- 1.5*IQR
upper = Q3 + 1.5*IQR
df = df[(df['reading score']> lower) & (df['reading score'] < upper)]
```

```python
Q1 = df['writing score'].quantile(0.25)
Q3 = df['writing score'].quantile(0.75)
IQR = Q3-Q1
lower = Q1- 1.5*IQR
upper = Q3 + 1.5*IQR
df = df[(df['writing score']> lower) & (df['writing score'] < upper)]
```

```python
fig , axs = plt.subplots(1,3,figsize = (15,12))

sns.boxplot(data = df['math score'], ax = axs[0], color = 'r')
sns.boxplot(data = df['reading score'], ax = axs[1], color = 'g')
sns.boxplot(data = df['writing score'], ax = axs[2], color = 'b')

plt.tight_layout()
plt.show()
```

```python
print("Math skew ness befor", df['math score'].skew())
df['math score'] = np.sqrt(df['math score'])
print("Math skew ness after", df['math score'].skew())

print("reading  skew ness befor", df['reading score'].skew())
df['reading score'] = np.sqrt(df['reading score'])
print("reading skew ness after", df['reading score'].skew())

print("writing skew ness befor", df['writing score'].skew())
df['writing score'] = np.sqrt(df['writing score'])
print("writing skew ness after", df['writing score'].skew())
```
