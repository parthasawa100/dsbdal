```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
fig, axes = plt.subplots(2,2,figsize = (10,8))
sns.boxplot(ax = axes[0][0], data = df['SepalLengthCm'])
sns.boxplot(ax = axes[0][1], data = df['SepalWidthCm'])
sns.boxplot(ax = axes[1][0], data = df['PetalLengthCm'])
sns.boxplot(ax = axes[1][1], data = df['PetalWidthCm'])
```

```python
fig, axes = plt.subplots(2,2,figsize = (10,8))
sns.boxplot(ax = axes[0][0],y = 'SepalLengthCm', hue = 'Species', data = df)
sns.boxplot(ax = axes[0][1],y = 'SepalWidthCm', hue = 'Species',data = df)
sns.boxplot(ax = axes[1][0],y = 'PetalLengthCm' , hue = 'Species',data = df)
sns.boxplot(ax = axes[1][1],y = 'PetalWidthCm', hue = 'Species',data = df)
```

```python
fig, axes = plt.subplots(2,2,figsize = (10,8))
sns.histplot(ax = axes[0][0],x = 'SepalLengthCm', hue = 'Species', data = df)
sns.histplot(ax = axes[0][1],x = 'SepalWidthCm', hue = 'Species',data = df)
sns.histplot(ax = axes[1][0],x = 'PetalLengthCm' , hue = 'Species',data = df)
sns.histplot(ax = axes[1][1],x = 'PetalWidthCm', hue = 'Species',data = df)
```
