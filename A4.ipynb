```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from seaborn import set_theme
```

```python
sns.get_dataset_names()
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe()
```

```python
set_theme(style = 'darkgrid')
```

```python
plt.figure(figsize = (8,6))
sns.countplot(x='class', hue = 'class', data = titanic, legend= False)
plt.title("Class VS Count")
plt.show()
```

```python
plt.figure(figsize= (8,6))
sns.countplot(x='class', hue='sex',data = titanic)
plt.title("Sex(Male, Female) Count IN Class")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.boxplot(x = 'class', y = 'age', hue = 'class', data = titanic, legend = False)
plt.title('Age Distribution in each class')
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.countplot(x = 'survived', data = titanic)
plt.title("Survived and Unsurvived Count")
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(x = 'survived', hue = 'class', data = titanic)
plt.title('survived as per class')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.countplot(x = 'survived', hue = 'sex', data = titanic)
plt.title('sex vs survived')
plt.show()
```

```python
plt.figure(figsize = (8,6))
sns.histplot(data = titanic['fare'], bins = 30, kde = True)
plt.show()
```
