```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
csv_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
col_names = ['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width','Class']
df =  pd.read_csv(csv_url, names = col_names)
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
fig, axes = plt.subplots(2,2,figsize = (18,15))

sns.histplot(ax = axes[0,0], x ='Sepal_Length', hue = 'Class',  data = df)
sns.histplot(ax = axes[0,1], x ='Sepal_Width', hue = 'Class',data = df)
sns.histplot(ax = axes[1,0], x ='Petal_Length', hue = 'Class',data = df)
sns.histplot(ax = axes[1,1], x ='Petal_Width', hue = 'Class',data = df)

plt.show()
```

```python
fig, axes = plt.subplots(2,2,figsize = (18,15))

sns.boxplot(ax = axes[0,0], x ='Sepal_Length', hue = 'Class',y ='Class',  data = df)
sns.boxplot(ax = axes[0,1], x ='Sepal_Width', hue = 'Class', y ='Class', data = df)
sns.boxplot(ax = axes[1,0], x ='Petal_Length', hue = 'Class', y ='Class',data = df)
sns.boxplot(ax = axes[1,1], x ='Petal_Width', hue = 'Class',y ='Class',data = df)

plt.show()
```
